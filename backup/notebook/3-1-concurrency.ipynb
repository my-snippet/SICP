\* SICP에 나오는 '병행성'에 대한 부분을 토대로 작성하였습니다. 운영체제에서 접하는 문제와 다를 수 있습니다.

예전에 어떤 친구가 __'쉬운 프로그램을 어렵게 만드려면, 여럿이서 같이 만들면 된다'__고 농담 같은 사실을 이야기하던 기억이 납니다.

알게 모르게 들어온 '사이좋게 같이'라는 미덕을 여기에 적용하면 __재앙__이 일어난다는 사실을 해본 사람은 모두 알고 있습니다.

이런 __재앙__때문에 팀 프로젝트는 웃으면서 시작해서 싸우면서 끝이나곤 합니다. 결국 '사이좋게 같이' 하지 못하고 다같이 망하거나, 한 명이 모두를 끌고 가는 행태를 너무도 많이 보고 겪었습니다.

팀원이 4명만 되어도 프로젝트 진행의 복잡도가 꽤 커집니다. 몇가지 예를 들어보겠습니다.

### 1. 4명이 해야할 일이 사이좋게 4가지라면?

### 2. 어떤 팀원이 하는 일이 다른 팀원이 해야할 일에 영향을 준다면?

### 3. 다른 일을 끝내야지만 이어서 진행할 수 있는 일이 있다면?

1.의 경우는 각자 사이좋게 일 하나씩 맡아서 처리하면 땡(?)인 것 같지만 안 그럴 때가 많습니다.

각자의 일인 것 같아도 수행과정에서 서로의 일에 영향을 주거나 서로의 진행상황이 필요한 경우가 많습니다.

( 일의 진행순서는 경우에 따라 $4! = 24$가지입니다. )

결국 일의 진행을 위해 상호간에 '대화'를 해야만 하는 상황이 일어납니다.

이 단계에서 '대화'를 잘 못하면 다같이 망하는 것 같습니다.

2.의 경우처럼 어떨 때는 자기가 하던 일을 '다른 일' 때문에 어느 정도 엎고 다시 시작해야합니다.

이와 같이 노동의 보람이 사라지는 순간이 반복되면 점점 의욕을 잃을지도 모릅니다.

3.의 경우처럼 일의 순서가 정해져있다면 좀 더 수월할 것입니다.

하지만 앞서 할 일이 좀처럼 끝나지 않을때, 뒤에 할 일을 기다리는 쪽은 계속 기다려야 하는 상황에 빠질 수 있습니다.

이렇게 기다리는 시간이 길어지면 가장 답답해하는 한 명이 모두를 끌고 가는 일이 벌어질 수 있습니다.

4명이서 4개의 일을 하는 것도 이렇게 쉽지 않습니다. 

__사람들이 이런 문제를 겪듯 프로그램 내에서도 비슷한 현상__이 일어날 수 있습니다.

프로그램이 사람 일을 흉내내어 옮겨놓았다면 이 것 역시 만만치 않을 것입니다.

1.의 경우처럼 각자의 일을 동시에 진행하는 것을 동시성Concurrency이라고 부릅니다. 이와 관련된 비유는 '[철학자의 식사](https://ko.wikipedia.org/wiki/%EC%8B%9D%EC%82%AC%ED%95%98%EB%8A%94_%EC%B2%A0%ED%95%99%EC%9E%90%EB%93%A4_%EB%AC%B8%EC%A0%9C)'가 유명합니다.

2.의 경우처럼 서로의 작업이 영향을 주는 것을 방지하도록 하는 것을 직렬화Serialization라고 합니다. 프로시저를 직렬화하면 어떤 일(프로세스)을 하려고 접근할 때 Mutex에게 물어보고 허락을 받아야지만( 다른 데서 이와 관련된 일을 수행하고 있지 않을 때 ) 일을 획득할 수 있습니다. 여기서는 '일의 어느 부분까지를 직렬화할 것인가'가 중요한 문제인 것 같습니다.

3.의 경우처럼 앞서 할 일이 끝나기를 계속 기다리는 것을 교착 상태Deadlock라고 부릅니다.
