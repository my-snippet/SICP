소개
===

이 문서는 SICP를 다룹니다. 책에 나온 개념과 예제를 다루며, 때로는 새롭게 문제를 정의해서 풀 예정입니다.  
SICP의 공식언어는 Scheme이지만, Python, Haskell을 이용하여 문제를 풀어보기도 할 것입니다.

책 표지 뒷면에 크게 적혀있는 ‘컴퓨터과학 입문교과서'라는 말이 있지만, 이게 진정 입문서라면 대다수가 그  
문을 열어보지도 못한 채 돌아서야 했을 것입니다. 다행히도 제가 다니는 학교의 제가 하고있는 전공에서는 이  
책을 입문서로 가르치지 않고, 다른 어느 과목에서도 가르치지 않습니다. 물론, 이 책은 대단한 책임에 틀림이  
없습니다. 그렇지만 지금까지 가르쳐주지 않아준 것에 감사하고 있습니다. 그 덕분에 지금 이 책을 보고 있는 것   
같습니다. 몇년 전에 어느 훌륭한 교수님이 ‘아, 이 좋은 입문서를 우리 학교 신입생들에게도 가르쳐야 겠다'고  
생각하고 이미 추진하지 않아서 다행이라는 말입니다. 그랬다면 프로그래밍은 일찌감치 포기했을 것입니다.

프로그래밍 이야기를 하기에 앞서 다른걸 생각해보았습니다. 먼저 여러 사람들이랑 일을 하거나 여행을 가거나  
하는 경우를 생각해보았습니다. 두 경우 모두 큰 사고만 안 치면 일은 나름 돌아가고 여행도 그럭저럭 할 수  
있습니다. 그걸로 만족할 수 있습니다. 그렇지만 일처리를 참 잘하는 사람이나 여행 경험이 많은 사람이랑 함께  
해본다면 생각이 달라집니다. 그런 사람과 함께하면 8시간 걸리던 일이 1시간 만에 끝나고, 혼나기만 하다가  
칭찬을 듣고, 길을 헤매는 대신 풍경을 누릴 수가 있습니다. 즉, 그들과 함께하면 ‘**더 효과적으로**’ 할  
수가 있습니다.

마찬가지로 똑같은 기능을 만들어도 더 효과적인 프로그램을 만들 수가 있습니다. 물론 ‘그런 프로그램을 만들 수  
있는 방법이 이 책이다'는 아닙니다. 그렇지만 ‘그 방법 중 하나가 이 책’일 수는 있을 것 같습니다.

자비스가 밝히는 고차원 프로시저의 유용성
===

영화 ‘아이언맨’을 떠올려보았습니다. 만약 자비스가 말만 할 줄알고 멍청했다면 토니 스타크'를 귀찮게 했을  
것입니다. 예를 들어, 토니스타크가 구해야 하는 신호가 있다고 가정합니다. 근데 그 함수가 너무 들쑥날쑥했기  
때문에 자비스에게 신호 함수를 다듬으라고(smoothing) 명령합니다.

> “자비스, 이 f에 smoothing을 n번 해둬.”

이 때, 자비스가 똑똑하다면 말이 끝나기가 무섭게 연산 결과를 말해주겠지만 멍청한 자비스라면 먼저,  
smoothing이 뭔지 물어볼 것입니다. 여기에 토니스타크가 'smoothing이란 $f(x_n)$를 그 주변 값  
( $f(x_{n-1})$, $f(x_{n+1})$ )들과 평균한 것이라고 가르쳐 줄 수 있을 것입니다. 하지만 자비스가 멍청하다면  
'그렇다면 평균은 무엇입니까?'하고 물어볼 것입니다. 이것은 여러번 주인을 귀찮게 해야 답을 구할 수가 있다는  
뜻입니다.

이렇듯 smoothing이라는 하나의 명령은 여러 하부 절차sub procedure로 이루어져 있는 고차원 프로시저higher  
-order procedure입니다. 고차원 프로시저를 잘 만들면 사람을 귀찮게 안하고 원하는 일을 손쉽고 빠르게 할 수가  
있습니다. 영화 속에 자비스는 고차원 프로시저로 가득할 것입니다.

n번째 수 찾기, 이름 없는 함수 이용하기(lambda), 황금비 구하기, 돈을 여러 동전으로 바꾸기, 지수 함수  
만들기, 지수 함수 값 빠르게 구하기, 이분법을 이용한 방정식의 해 찾기, 미분하기, 적분하기(simpson 공식
